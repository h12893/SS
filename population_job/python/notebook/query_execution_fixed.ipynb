```python
# 実行するクエリを定義
query = """
SELECT
    *
FROM economic_census_employments_curated
"""

print(f"実行するクエリ:")
print(query)
```

```python
# クエリを実行してデータを取得
df = fetch_data(query)
print(f"✓ クエリ実行完了")
print(f"取得行数: {len(df)}")
print(f"取得列数: {len(df.columns)}")
```
