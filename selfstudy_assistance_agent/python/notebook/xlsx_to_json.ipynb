## スキルチェック（V.1.1.0）

### 指定した名前のファイルをinputディレクトリからデータフレームで読み込み

```python
file_name = 'skillcheck_ver6.00.xlsx'
file_path = os.path.join(data_dir_input, file_name)
sheet_name_dict = {'基盤':'base', '価値創造力':'value', 'データサイエンス力':'DS', 'データエンジニアリング力':'DE', '融合':'fusion'}
data_dict = {}
column_check_list = []
for key, value in sheet_name_dict.items():
    rename_dict = {}
    data_dict[value] = Read_xlsx_to_df(f'{file_path}', 
                                       sheet_name=key, 
                                       header=2, 
                                       reset_index=True)
    print(data_dict)
    for column in data_dict[value].columns:
        rename_dict[column] = re.sub(r'[\r\n]+', '', column).replace(" ", "")
        column_check_list.append(rename_dict[column])
    data_dict[value] = data_dict[value].rename(columns=rename_dict)

column_check_list = list(set(column_check_list))
print("列名の確認:")
for column in column_check_list:
    print(column)
```

### データフレーム内容を辞書にまとめてJSON形式で出力

```python
skill_1_list = ["base", "DS", "DE", "fusion"]

total_dict = {}
for key, value in sheet_name_dict.items():

    sheet_list = []
    if value in skill_1_list:

        for number in data_dict[value]['No'].unique():
            column_list = [column for column in data_dict[value].columns if column != 'No']
            individual_dict = {}
            for column in column_list:
                if number == '–':
                    individual_dict['No'] = number
                else:
                    individual_dict['No'] = int(number)

                if (number != '–') and (column == 'SubNo'):
                    contents = int(data_dict[value].query(f'No=={number}')[column][:1].values[0])
                elif (number != '–') and (column != 'SubNo'):
                    contents = data_dict[value].query(f'No=={number}')[column][:1].values[0]
                else:
                    contents = data_dict[value].query("No=='–'")[column][:1].values[0]
                individual_dict[column] = contents
            sheet_list.append(individual_dict)

    else:

        for number in data_dict[value]['No'].unique():
            column_list = [column for column in data_dict[value].columns if (column != 'No') and (column != 'SubNo')]
            for step in range(1, 4):
                individual_dict = {}
                for column in column_list:
                    individual_dict['No'] = int(number)
                    individual_dict['SubNo'] = int(data_dict[value].query(f'No=={number}')['SubNo'][:1].values[0])
                    individual_dict['STEP'] = step
                    if (column not in ["★（見習い）", "★★（一人前）", "★★★（棟梁）"]):
                        individual_dict[column] = data_dict[value].query(f'No=={number}')[column][:1].values[0]
                    else:
                        if (step == 1):
                            individual_dict["スキルレベル"] = "★" 
                            individual_dict["チェック項目"] = data_dict[value].query(f'No=={number}')["★（見習い）"][:1].values[0] 
                        elif (step == 2):
                            individual_dict["スキルレベル"] = "★★" 
                            individual_dict["チェック項目"] = data_dict[value].query(f'No=={number}')["★★（一人前）"][:1].values[0] 
                        elif (step == 3):
                            individual_dict["スキルレベル"] = "★★★" 
                            individual_dict["チェック項目"] = data_dict[value].query(f'No=={number}')["★★★（棟梁）"][:1].values[0] 
                    
                sheet_list.append(individual_dict)


    total_dict[value] = sheet_list
```
